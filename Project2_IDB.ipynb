```python
print("Data info:")
df2.info()
print("\nMissing values:")
print(df2.isnull().sum())
```

```python
df2 = df2.drop(columns=['2027', '2028', '2029', '2030', '2031', '2032'])
```

```python
df2.info()
```

```python
year_cols = [col for col in df2.columns if str(col).isdigit()]
id_cols = [col for col in df2.columns if col not in year_cols]

df2_long = df2.melt(
    id_vars=id_cols,
    value_vars=year_cols,
    var_name='Year',
    value_name='Value'
)

df2_long['Year'] = df2_long['Year'].astype(int)
display(df2_long)
```

```python
df2_long.info()
```

```python
df2 = df2.drop(columns=['2025','2026'])
```

```python
df2.isnull().sum()
```

```python
df2_long = df2_long[~df2_long[id_cols].isna().all(axis=1)].copy()

df2_long['Value'] = pd.to_numeric(df2_long['Value'], errors='coerce')

valid_years = list(range(2000, 2027))
df2_long = df2_long[df2_long['Year'].isin(valid_years)].reset_index(drop=True)

print("After cleaning:")
print(df2_long.info())
print("\nMissing values:")
print(df2_long.isnull().sum())
```

```python
display(df2_long)
```

```python
a=df2_long.groupby("Counterpart-Area Code").agg("count")
#df2_long.apply(display)
display(a)
```

```python
#df2_long.info()
df2_long.head()
```

```python
x=df2_long.groupby('Country Name').agg('count')
display(x)
```

```python
threshold = (df2_long.shape[1] + 1) // 2
df2_long = df2_long.dropna(thresh=threshold).reset_index(drop=True)

print(f"Kept rows with at least {threshold} non-NaNs.")
df2_long.info()
```

```python
df2_long.isnull().sum()
```

```python
df2_long= df2_long.dropna(subset=['Value'])
```

```python
# Check the distribution of null values by country
nulls_by_country = df2_long.groupby('Country Code')['Value'].apply(lambda x: x.isnull().sum())
print("Countries with null values:")
print(nulls_by_country[nulls_by_country > 0].sort_values(ascending=False))

# Check how many non-null values per country
non_nulls_by_country = df2_long.groupby('Country Code')['Value'].apply(lambda x: x.notna().sum())
print("\nNon-null values per country:")
print(non_nulls_by_country.sort_values())

# Check the original data before conversion to numeric
print("\nValue column data type:")
print(df2_long['Value'].dtype)
print("\nSample of values:")
print(df2_long[['Country Code', 'Series Name', 'Year', 'Value']].head(20))
```
